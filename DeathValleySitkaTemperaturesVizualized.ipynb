```python
"""This is a visualization of low and high temperatures in Death Valley, CA. 
It is one of the exercises from the book Python Crash Course"""

import csv
import matplotlib.pyplot as plt
from datetime import datetime

filename_1 = 'death_valley_2018_simple.csv'
with open(filename_1) as f_1:
    reader_1 = csv.reader(f_1)
    header_row_1 = next(reader_1)
    # Get dates, and high and low temperatures from this file.
    highs_DV, dates_DV, lows_DV = [], [], []
    for row_1 in reader_1:
        current_date_1 = datetime.strptime(row_1[2], '%Y-%m-%d')
        try:
            high_1 = int(row_1[5])
            low_1 = int(row_1[6])
        except ValueError:    
            print(f'Missing data for {current_date_1}')
        else:
            dates_DV.append(current_date_1)
            highs_DV.append(high_1)
            lows_DV.append(low_1)
            
filename_2 = 'sitka_weather_2018_simple.csv'
with open(filename_2) as f_2:
    reader_2 = csv.reader(f_2)
    header_row_2 = next(reader_2)
    # Get dates, and high and low temperatures from this file.
    highs_S, dates_S, lows_S = [], [], []
    for row_2 in reader_2:
        current_date_2 = datetime.strptime(row_2[2], '%Y-%m-%d')
        try:
            high_2 = int(row_2[5])
            low_2 = int(row_2[6])
        except ValueError:    
            print(f'Missing data for {current_date_2}')
        else:
            dates_S.append(current_date_2)
            highs_S.append(high_2)
            lows_S.append(low_2)            
    
    
    
    # Plot the low and high temperatures.
    
    plt.style.use('seaborn')
    fig, ax = plt.subplots(figsize=(15,15))
    
    
    ax.plot(dates_DV, highs_DV, c='red', alpha=0.5, label='highs in DV')
    ax.plot(dates_DV, lows_DV, c='blue', alpha=0.5, label='lows in DV')
    ax.plot(dates_S, highs_S, c='orange', alpha=0.5, label='highs in Sitka')
    ax.plot(dates_S, lows_S, c='#102E4A', alpha=0.5, label='lows in Sitka')
    plt.fill_between(dates_DV, highs_DV, lows_DV, facecolor='#0D4F89', alpha=0.1)
    plt.fill_between(dates_S, highs_S, lows_S, facecolor='#7D2E98', alpha=0.1)
    
    # Format plot.
    
    plt.title('Daily high and low temperatures - 2018 \nDeath Valley, CA \nSitka, AK', fontsize=24)
    plt.xlabel('', fontsize=16)
    fig.autofmt_xdate()
    plt.ylabel('Temperature (F)', fontsize=16)
    plt.tick_params(axis='both', which='major', labelsize=16)
    plt.ylim(10,130)
    
    plt.legend()
    plt.show()
```
